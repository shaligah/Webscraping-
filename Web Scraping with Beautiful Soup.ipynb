```python
#importing the necessary libraties
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#extracting the data from the website
url = 'https://www.formula1.com/en/results/2024/drivers'
response = requests.get(url)
content = response.content
```

```python
#parsing the html
parser = BeautifulSoup(content,'html.parser')
F1table = parser.find('table')
```

```python
headers = []
for i in F1table.find_all('th'):
    headers.append(i.get_text().strip())
```

```python
rows = []
for i in F1table.find_all('tr')[1:]:
    r = i.find_all('td')
    rd = [data.get_text().strip() for data in r]
    rows.append(rd)
```

```python
#creating a dataframe from the data 
df = pd.DataFrame(rows, columns=headers)
```
